## if_elif_else

### 1. AGE

```python
age = 20
if age < 18:
    print("Minor")
elif age < 65:
    print("Adult")
else:
    print("Senior")
```

### 2. Grading system

```python
score = 85 
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")
```

### 3. Temperature Check


```python
temp = 10
if temp > 30:
	print("Hot")
elif temp > 15:
	print("Warm")
else:
	print("Cold")
```

### 4. Number sign

```python
num = -5
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

## 5. Time of Day

```python
hour = 14
if hour < 12:
    print("Morning")
elif hour < 18:
    print("Afternoon")
else:
    print("Evening")
```

### 6. Speed Alert

```python
speed = 75
if speed > 100:
    print("Too Fast")
elif speed > 60:
    print("Normal Speed")
else:
    print("Slow")
```

#### Six Example Of Match Case

## 1. Day of the Week

```python
day = "Monday"
match day:
    case "Saturday" | "Sunday": print("Weekend!")
    case _: print("Work day")
```

## 2. HTTP Status Codes 

```python
code = 404
match code:
    case 200: print("Success")
    case 404: print("Not Found")
    case 500: print("Server Error")
    case _: print("Unknown Code")
```

## 3. calculations

```python
op = "+"
match op:
    case "+": print("Addition")
    case "-": print("Subtraction")
    case "*": print("Multiplication")
    case _: print("Invalid Op")
```
